```python
import os 
from dotenv import load_dotenv
load_dotenv()

os.environ["HUGGINGFACEHUB_API_TOKEN"] = os.getenv("HF_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")

from langchain.chat_models import init_chat_model

model = init_chat_model(
    "llama-3.1-8b-instant",
    model_provider="groq"
)
```

# Text Prompt

```python
response = model.invoke("What is the waeather in mumbai")
response
```

# Message Prompts

1. System Message
2. Humans Message
3. AI Message
4. Tool Message

```python
from langchain.messages import SystemMessage , AIMessage, HumanMessage ,ToolMessage


messeges =[
    SystemMessage("You are a poetry expert"),
    HumanMessage("Give me in one poem in 10 words")

]

response  = model.invoke(messeges)

print(response.content)
```

```python
human_msg  = HumanMessage(
    content='Hello',
    name="nithin",
    id = "msg_123"
)
```

```python
response  = model.invoke([human_msg])

print(response.content)
```

```python
message = [
    SystemMessage("Your are react on the Situations shared"),
    HumanMessage("Can you please help on this") ,
    AIMessage("Please Share the Situation"), 
    HumanMessage("They are Physical  fighting") ]

response  = model.invoke(message)

print(response.content)
```

# Tools Messages

```python
ai_msg =AIMessage(
    content =[],
    tool_calls =[{
        'name':"get_weather",
        'args':{'location':'San Francisco'},
        "id":"call_123"
    }]
)



weather ="sunny 72F"

tool_message =ToolMessage(
    content= weather, 
    tool_call_id ="call_123"
)

message =[
    HumanMessage("waht is the weather in San Franciso"),
    ai_msg,
    tool_message
]

response  = model.invoke(message)

print(response.content)
```
