```python
import os 
from dotenv import load_dotenv
load_dotenv()

os.environ["LANGCHAIN_API_KEY"] = os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGCHAIN_PROJECT"] = os.getenv("LANGCHAIN_PROJECT")
groq_api_key= os.getenv("GROQ_API_KEY")

```

```python
from langchain_groq import ChatGroq
from langchain_openai import ChatOpenAI

model = ChatGroq(model= 'moonshotai/kimi-k2-instruct', groq_api_key =groq_api_key)

model
```

```python
from langchain_core.messages import  HumanMessage , SystemMessage

messages =[
    SystemMessage(content='Give me the beautile pickupline to girl '),
    HumanMessage(content='kokile')
]

result = model.invoke(messages)
```

result

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()

parser.invoke(result)
```

```python
chain = model |parser 
chain.invoke(messages)
```

#  Prompt Templates

```python
from langchain_core.prompts import ChatPromptTemplate  


generic   = "Give me the Good pickup lines"

ChatPrompt = ChatPromptTemplate.from_messages(
    [("system",generic),("user","{text}")]
)

```

```python
result = ChatPrompt.invoke({"text":"Priya"})
```

```python
result.to_messages()
```

```python
chain = ChatPrompt |model |parser

response =chain.invoke({"text":"Prathi"})
print(response)
```
