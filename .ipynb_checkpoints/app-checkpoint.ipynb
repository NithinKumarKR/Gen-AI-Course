```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
df = pd.read_csv("insulin_dosage_prediction.csv")
```

```python
df.columns
```

```python
df.drop(['patient_id'],axis=1,inplace=True)
```

```python
df.head()
```

```python
gender_label =LabelEncoder()
df['gender']=gender_label.fit_transform(df['gender'])
family_history_label =LabelEncoder()
df['family_history']=gender_label.fit_transform(df['family_history'])
Insulin_label =LabelEncoder()
df['Insulin']=Insulin_label.fit_transform(df['Insulin'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
one = OneHotEncoder()
food_intake_one =one.fit_transform(df[['food_intake']])
food_intake_one_pd =pd.DataFrame(food_intake_one.toarray(), columns=one.get_feature_names_out(['food_intake']))
df = pd.concat([df, food_intake_one_pd], axis=1)

previous_medications_label = OneHotEncoder()
previous_medications =previous_medications_label.fit_transform(df[['previous_medications']])
exercise_one_pd =pd.DataFrame(previous_medications.toarray(), columns=previous_medications_label.get_feature_names_out(['previous_medications']))
df = pd.concat([df, exercise_one_pd], axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['food_intake', 'previous_medications'],axis=1,inplace=True)
```

```python
# Save the Pickle File
with open("gender_label.pkl","wb") as f:
    pickle.dump(gender_label,f)
with open("family_history_label.pkl","wb") as f:
    pickle.dump(family_history_label,f)
with open("Insulin_label.pkl","wb") as f:
    pickle.dump(Insulin_label,f)
    
# Save the Pickle File
with open("food_intake_one.pkl","wb") as f:
    pickle.dump(previous_medications,f)
with open("previous_medications.pkl","wb") as f:
    pickle.dump(previous_medications,f)
```

```python
x = df.drop('Insulin', axis=1)
y = df['Insulin']
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=42)
```

```python
x.columns
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
# Transform x_test using the same scaler
x_test_scaled = scaler.transform(x_test)
```

```python
with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)
```

# ANN

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping , TensorBoard
import datetime
```

```python
model =Sequential(
    [
        Dense(64 , activation='relu',input_shape =(x_train.shape[1],)) , # HL1
        Dense(32 , activation='relu') , # HL2
        Dense(4 , activation='softmax') # OP
    ]
)
```

```python
model.summary()
```

```python
opt =tf.keras.optimizers.Adam(learning_rate=.01)
loss =tf.keras.losses.BinaryCrossentropy()
```

```python
# Compile the model
model.compile(optimizer=opt,loss = "sparse_categorical_crossentropy", metrics=['f1_score'])
```

```python
# Set the Tensorboard

log_dir = 'logs/fit/'+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensor_flow_callback =TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
# Set Up Early Stopping 
early_stopping_callback =  EarlyStopping(monitor='val_loss', patience=50,  restore_best_weights=True)
```

```python

history = model.fit(
    x_train_scaled, y_train,
    validation_data=(x_test_scaled, y_test),
    epochs=1000,
    callbacks=[tensor_flow_callback, early_stopping_callback]
)
```

```python
model.save('model.h5')
```

```python
import warnings
from tensorflow.keras.models import load_model
import pickle
import numpy as np
model = load_model('model.h5')
# Load encoders
with open('exercise_label.pkl','rb') as file:
    exercise_encoder = pickle.load(file)
with open('gender_label.pkl','rb') as file:
    gender_encoder = pickle.load(file)
with open('member_one.pkl','rb') as file:
    member_encoder = pickle.load(file)

input = {'Age':30,'Gender':'Male','Favorite_Exercise':'Squats','Avg_Workout_Duration_Min':60,'Avg_Calories_Burned':10,'Total_Weight_Lifted_kg':20,
'Visits_Per_Month':2,'Membership_Type':'Quarterly','Period':100}

# Label encoded features
gender_enc = gender_encoder.transform([input['Gender']])[0]
member_enc = member_encoder.transform([[input['Membership_Type']]]).toarray()[0]
exercise_enc = exercise_encoder.transform([[input['Favorite_Exercise']]]).toarray()[0]

print("Gender Encoded:", gender_enc)
print("Exercise Encoded:", exercise_enc)
print("Membership Encoded:", member_enc)
```

```python
feature_names = member_encoder.get_feature_names_out()
encoded_row = member_encoder.transform([[input['Membership_Type']]]).toarray()[0]
member = {feature_names[i]: int(encoded_row[i].item()) for i in range(len(feature_names))}
feature_names = exercise_encoder.get_feature_names_out()
encoded_row = exercise_encoder.transform([[input['Favorite_Exercise']]]).toarray()[0]
exercises = {feature_names[i]: int(encoded_row[i].item()) for i in range(len(feature_names))}
```

```python
input.pop('Favorite_Exercise')
input.pop('Membership_Type')
input['Gender'] = int(gender_enc)
input.update(member)
input.update(exercises)
input['Favorite_Exercise_Bench_Press']=input['Favorite_Exercise_Bench Press']
input['Favorite_Exercise_Pull_ups']=input['Favorite_Exercise_Pull-ups']
input.pop('Favorite_Exercise_Bench Press')
input.pop('Favorite_Exercise_Pull-ups')
```
