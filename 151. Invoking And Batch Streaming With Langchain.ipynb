# Streaming

```python
import os 
from dotenv import load_dotenv
load_dotenv()

os.environ["HUGGINGFACEHUB_API_TOKEN"] = os.getenv("HF_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")


```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(
    "llama-3.1-8b-instant",
    model_provider="groq"
)
```

```python
for i in model.stream('Give me the 10 picklines for girls'):
    print(i.text,end='|',flush=True)
```

```python
for i in model.stream('Give me the 10 picklines for girls'):
    print(i.text,end='*',flush=True)
```

# Batching

```python
%%time
responses  = model.batch([
    "Applications of AI in 10 words",
    "Applications of cyber security in 10 words",
    "Tell one best line in 10 words"
], config = {'max_concurrency':1,}
)

for response in responses:
    print(response.content)
```
